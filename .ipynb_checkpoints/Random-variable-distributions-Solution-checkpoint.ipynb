```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the dataframe
df = pd.read_csv('./files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
df.head(2)
```

# 1. Get the numerical variables from our dataset.

```python
numerical = df.select_dtypes('number')
numerical.head(2)
```

# 2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.

```python
def drawplots(df):
    for c in df.columns:
        sns.displot(data = df[c], bins = 100)
        plt.show()
```

```python
drawplots(numerical)
```

```python
# Customer Lifetime Value : Exponential
# Income : Normal, skewed to the left, many 0 outliers
# Monthly Premium Auto: Exponential
# Month since last claim: Conmtinuous distribution
# Month since policy inception: Normal (very flat)
# Number of open complaints: Exponential
# Number of policies: Exponential
# Total claim amount: Exponential
```

# 3. Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

### 3.1 Log10

```python
# log10 transformation
def log_transfom_clean(x):
    x = np.log10(x)
    if np.isfinite(x):
        return x
    else:
        return 0
```

```python
numerical_log = pd.DataFrame([])
for c in numerical.columns:
    numerical_log[c] = numerical[c].apply(log_transfom_clean)
```

```python
drawplots(numerical_log)
```

```python
# Customer Lifetime Value and total claim amount now look more like a normal distribution.
```

### 3.2 MinMax Scaler

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
transformer = MinMaxScaler().fit(numerical)
X_normalized = transformer.transform(numerical)
X_normalized = pd.DataFrame(X_normalized, columns=numerical.columns)
```

```python
drawplots(X_normalized)
```

```python
# Month since policy inception now got better!
```

### 3.2 MinMax Scaler

```python
from sklearn.preprocessing import StandardScaler
transformer = StandardScaler().fit(numerical)
X_normalized = transformer.transform(numerical)
X_normalized = pd.DataFrame(X_normalized, columns=numerical.columns)
```

```python
drawplots(X_normalized)
```

```python
# The results of the Standard Scaler look very similar to the results of the MinMax Scaler
```

### 3.3 RobustScaler

```python
# Trying out another scaler from sklearn
from sklearn.preprocessing import RobustScaler
```

```python
# Simply speaking in contrast to the StandardScaler which uses the median and
# the standard deviation for scaling,
# the RobustScaler calculates with the median and the interquartile range,
# that has the effect, that like the median, the RobustScaler is not very prone
# or "Robust" against outliers, we look if we see another effect here.
```

```python
transformer = RobustScaler().fit(numerical)
```

```python
X_normalized = transformer.transform(numerical)
X_normalized = pd.DataFrame(X_normalized, columns=numerical.columns)
```

```python
drawplots(X_normalized)
```

```python
# With this method we see some signs of normalization on Customer Lifetime Value
# and Months since last claim
```
